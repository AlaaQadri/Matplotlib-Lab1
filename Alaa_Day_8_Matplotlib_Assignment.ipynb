```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
X = [2019,2020,2023,2024]
plt.plot(X)
```

```python
Y = {'Apple': 40 , 'Banana': 30, 'Orange': 60}
data = pd.Series(Y)
plt.plot(data)
```

```python
data = {'SalesPerson': ['Kathey', 'Michael', 'William', 'Kathey', 'William', 'Kathey', 'Michael'],
        'Region': ['East', 'West', 'North', 'South', 'North', 'North', 'East'],
        'OrderAmount': [600, 700, 400, 500, 400, 700, 800]}
df = pd.DataFrame(data, columns=['SalesPerson', 'Region', 'OrderAmount', 'Month', 'isAccepted'])

plt.hist(df['SalesPerson'], facecolor = 'b', edgecolor = 'red', bins = 3)
plt.xlabel('SalesPerson Average')
plt.ylabel('OrderAmount')
plt.title('Order amount for each person')

plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
fruite = ["Apples", "Bananas", "Cherries", "Dates"]
freq = np.array([35, 25, 25, 15])

plt.pie(freq,labels=fruite, autopct='%1.1f%%', startangle=90, shadow=True,
        radius=1, explode = [0.3,0.3,0.3,0.3])
plt.show() 
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df['SalesPerson'],df['OrderAmount'], marker = 'o', s = 50, alpha = 0.4)
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(df['SalesPerson'],df['OrderAmount'], marker = '*', s = 50, label = 'Order Amount', alpha = 0.4)
plt.legend(loc = 'upper center')
plt.title('salesperson vs. order amount',fontsize = 20)
plt.xlabel('Sale person',fontsize = 16)
plt.ylabel('Order Amount',fontsize = 16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
from matplotlib import colors

plt.plot(range(1960,2011,2), c='red',label='1')
plt.plot(range(1970,2011,3), c='green',label='2')
plt.plot(range(1960,2013,1), c='blue',label='3')
plt.legend( shadow = True, loc = 'upper left') 
plt.show()
```

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 20)

#first
plt.subplot(1,3,1) 
plt.pie(freq,labels=fruite, autopct='%1.1f%%')
plt.title('Fruits');

# second
plt.subplot(1,3,2)
plt.scatter(df['SalesPerson'],df['OrderAmount'])
plt.title('SalesPerson vs. OrderAmounte')
plt.xlabel('Sales Person')
plt.ylabel('Order Amount'); 

# #third
plt.subplot(1,3,3)
plt.scatter(df['SalesPerson'],df['OrderAmount'])
plt.title('SalesPerson vs. OrderAmount')
plt.xlabel('Sales Person')
plt.ylabel('Order Amount'); 
```
